## Function

```python
def add2(x,y) :
  n = x + y
  return n

def add3(x,y,z=5) :
  n = x + y + z
  return n
```

```python
print(add2(100,200), add3(100,200), add3(100,200,300))
```

```python
def test(x,y,z=1) :
  print("x = ",x)
  print("y = ",y)
  print("z = ",z)
  return
```

```python
test(100,200,300)
```

```python
test(z=100,y=200,x=300)
```

```python
test(y=200,x=300)
```

```python
def add_mul(x,y) :
  sum = x + y
  mul = x * y
  return sum,mul
```

```python
a = add_mul(100,200)
print(a,type(a))    # tuple로 받음
```

```python
b,c = add_mul(100,200)
print(b,c,type(b), type(c))
```

```python
n1, n2 = 100, 200

def add2(n1,n2) :  # 함수 선언 시 파라미터는 함수 안에서만 동작
  print(n1,n2)
  return n1+n2    # return받는 변수가 없어도 괜찮다.

add2(1,2)
print(n1,n2)
```

## The number of parameters are not fixed

```python
def test(*args) :
  print(type(args))
  print(args)
  for item in args :
    print(item)
```

```python
test()
test(1,2)
test(1,2,3,4)
```

```python
def test1(**args) :
  print(type(args))
  print(args)
  for key in args :      # dict의 key는 string으로 받아진다.
    print(key,args[key])

def test2(**args) :
  print(type(args))
  print(args)
  for key, value in args.items() :
    print(key,value)
```

```python
test1(a=1,b=2)
test1(a=1,b=2, c=3)
```

### format() / place holder : {}
    

```python
a = "온도 : {}, 습도 : {}%, 강수확률 : {}%".format(20,90,100)
print(a)
```

```python
a = "온도 : {temp}, 습도 : {humid}%, 강수확률 : {rain}%".format(humid =90, temp =20, rain = 100)
print(a)
```
