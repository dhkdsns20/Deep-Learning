### OpenCV 의 영상은 칼라순서가 RGB가 아니라 BGR로 되어 있음

````python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('test.jpg') 
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) # BGR -> RGB
plt.imshow(img)
plt.show()
````
